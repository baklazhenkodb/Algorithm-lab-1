```python
#Import all neseccarily libraries
import networkx as netx #Graph library
import sys
from collections import deque
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(threshold = sys.maxsize)
```

#### Generate random graph

```python
Graph = netx.generators.random_graphs.gnm_random_graph(100, 200, 2)

netx.draw(Graph,  with_labels=True, node_color='Red', width=1, edge_color='Gray', font_size=5, node_size=80)
plt.show()
```

#### Print adjacency matrix

```python
adj_matrix = netx.to_numpy_matrix(Graph)

adj_matrix[:4,:]
```

#### Print adjacency list

```python
adj_list = netx.convert.to_dict_of_lists(Graph)

for item in adj_list:
    print(f'Node {item}: {adj_list[item]}')
```

#### DFS

```python
start = 0
v_list = []
visited = [False] * (100)
print(visited)
v_list.append(start)

def dfs(adg_list, start):  
    
    visited[start] = True
    
    for v in adg_list[start]:
        if not visited[v]:
            dfs(adg_list, v)
            v_list.append(v)
            
dfs(adj_list, start)
print(f'DFS: {v_list}')

#Calculate connected components
visited = [False] * (100)
components = 0

for i in range(100):
    if not visited[i]:
        components += 1
        dfs(adj_list, i)

print(f'Number of connected components: {components}')
```

#### BFS

```python
#Breadth first search
def bfs (adg_list, first_vertex, second_vertex):
    
    visited_nodes=[]
    que = deque()
    que.append(first_vertex)
    flag = True
    
    while (que) and (flag):      
        
        current_vertex = que.popleft()
        
        if second_vertex != current_vertex:
            visited_nodes.append(current_vertex)
            
            for item in adg_list[current_vertex]:
                if item != second_vertex:   
                    if (item not in visited_nodes) and (item not in que):
                        que.append(item)
                else:
                    flag = False
                    visited_nodes.append(second_vertex)
                    print (f'The shortest way between [{first_vertex}] and [{second_vertex}]:')
                    print(visited_nodes)
                    break
                    
    if flag == True:
        print(f'Error, the node is unreachable' )
        
bfs(adj_list, 0, 99)
```
